```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#Sample data
#Independent variable (year of experience)
X = np.array([1,2,3,4,5,6,7,8]).reshape(-1,1)
#Depedent Variable (Salary in thousand)
y = np.array([45,50,60,65,70,75,80,85])
```

```python
#creating the linear regression model
model = LinearRegression()
```

```python
#Train the model
model.fit(X,y)
```

```python
#make a prediction for a new data
new_experience = np.array([[5]])
predicted_salary = model.predict(new_experience)
```

```python
print(f"Predicted salary for {new_experience[0][0]} years of experience: ${predicted_salary}")

#Plot the data and the regression line
plt.scatter(X,y, color ='blue',label='Actual Data')
plt.title('Salary vs Years of Experience')
plt.xlabel("Years of Experience")
plt.ylabel("Salary (in thousands)")
plt.legend()
plt.grid(True)
plt.show()
```
