```python
#importing the necessary libraries and datasets
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
```

```python
#load iris dataset from scikit-learn
iris = datasets.load_iris()
```

```python
#to make the visulaization easy lets use two features (petal length and petal width)
X = iris.data[:,2:]
y = iris.target
```

```python
#create and train the SVM model
C = 1.0
svc_model = svm.SVC(kernel='linear',C=C).fit(X,y)
```

```python
#Creating the decision boundries
x_min,x_max = X[:,0].min() - 1, X[:,0].max() + 1
y_min,y_max = X[:,0].min() - 1, X[:,0].max() + 1
#step the size for the mesh grid
h = 0.02
xx, yy = np.meshgrid(np.arange(x_min,x_max,h),
                     np.arange(y_min,y_max,h))

#using the training model to predict the class
Z = svc_model.predict(np.c_[xx.ravel(), yy.ravel()])

#reshape the prediction to match the mesh grid
Z = Z.reshape(xx.shape)
```

```python
#create the plot
plt.figure(figsize=(10,7))

plt.contourf(xx,yy,Z,cmap=plt.cm.coolwarm, alpha=0.8)

#Overlay the original data points on the plot
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.coolwarm,edgecolors='k')

#Adding the labels and titles
plt.xlabel('Petal length (cm)')
plt.ylabel('Petal width(cm)')
plt.title('SVM decision Boundries ')
plt.show()
```
