```python
#import the necessary libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
#Sample data
data = {
    'Age':[25,35,45,20,50,30,40,22],
    'Income': [ 'High','Low','High','Low','High','Medium','Medium','Low'],
    'Buys': [0,1,1,0,1,0,1,0] #0:No 1:Yes
}
df = pd.DataFrame(data)
```

```python
#convert the categorial into numerical
df['Income'] = df['Income'].map({'Low':0,'Medium':1,'High':2})
```

```python
#Features and the target variable
X = df[['Age','Income']]
y = df['Buys']
```

```python
#spliting the data into training and testing model
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
#Create a decision tree classifier
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
#Make the predicitons
y_pred = clf.predict(X_test)
```

```python
#Evaluate the accuracy
print(f"Accuracu:{accuracy_score(y_test,y_pred):.2f}")
```

```python
#Visualize the decision tree
plt.figure(figsize=(10,8))
tree.plot_tree(clf,filled=True,feature_names=['Age','Income'],class_names=['Does not Buy','Buy'])
plt.show()
```
