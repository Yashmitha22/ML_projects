```python
#import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#Creating the sample data
X = np.array([[5,3],[10,15],[15,12],[24,10],[30,10],[85,70],
              [71,80],[60,78],[70,55],[80,91]])
```

```python
#using the nested cluster in teh linkage method
linked = linkage(X , method='ward')
```

```python
#Plot the dendrogram
plt.figure(figsize=(10,7))
dendrogram(linked,orientation='top',
           labels=np.arange(1,11),
           distance_sort = 'descending',
           show_leaf_counts=True)

plt.title('Hierachical Clustering Dendrogram')
plt.xlabel('Data Point Index')
plt.ylabel('Distance')
plt.show()

cluster = AgglomerativeClustering(n_clusters=2,metric ='euclidean',linkage='ward')
labels = cluster.fit_predict(X)

print("Cluster labels for each data point(when specifying 3 clusters):")
print(labels)

#plot the graph to make teh concept clear
plt.figure(figsize=(10, 7))
plt.scatter(X[:,0],X[:,1],c = labels,cmap = 'rainbow')
plt.title("Data points clustered into 3 groups")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

```
