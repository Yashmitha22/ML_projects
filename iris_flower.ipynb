```python
#importing the necessary dataset and libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#Load the dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
#Splitting the data into training and testing sets
# It helps evaluate how well the model perfrom on new, unseen data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#Creating and training the classification tree model
model = DecisionTreeClassifier(max_depth=3, random_state =42)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
#Evaluate the model performance
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy on Test Data:{accuracy *100 :.2f}%")
print("---")

plt.figure(figsize=(20,10))
plot_tree(model,feature_names=iris.feature_names,class_names=iris.target_names,
          filled=True,
          rounded=True,
          fontsize=12)
plt.title("Decision tree for iris classification(max_depth=3)")
plt.show()
```
