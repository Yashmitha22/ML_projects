```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
#Generate the synthetic cluster using four cluster
X,y_true = make_blobs(n_samples = 300,centers=4,cluster_std=0.70,random_state=0)
```

```python
#plot the generated data
plt.scatter(X[:, 0],X[:, 1],s=20)
plt.title("Original Unclustered Data")
plt.show()
```

```python
#Initialize the K mean cluster
kmeans = KMeans(n_clusters=4, random_state=0, n_init=10)

kmeans.fit(X)

y_kmeans = kmeans.predict(X)

#get the coordinates of the cluster centers
centers = kmeans.cluster_centers_
#plotting the data points
plt.scatter(X[:, 0],X[:, 1],c=y_kmeans,s=20,cmap = 'viridis')
#ploting the centroids
plt.scatter(centers[:, 0],centers[:, 1],c='red',s=200,alpha=0.75,marker='*')

plt.title("K- mean clustering Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
