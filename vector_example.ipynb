```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
v1 = np.array([2, 3])
v2 = np.array([-1, 2])

# Plot vectors
plt.figure()
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1)

plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid()
plt.axhline(0)
plt.axvline(0)
plt.title("Vectors in 2D Space")
plt.show()
```

```python
# Vector addition
v3 = v1 + v2

plt.figure()
plt.quiver(0, 0, v1[0], v1[1], color='blue', scale=1)
plt.quiver(v1[0], v1[1], v2[0], v2[1], color='green', scale=1)
plt.quiver(0, 0, v3[0], v3[1], color='red', scale=1)

plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.grid()
plt.title("Vector Addition")
plt.show()
```

```python
# Dot product
dot = np.dot(v1, v2)

# Angle between vectors
cos_theta = dot / (np.linalg.norm(v1) * np.linalg.norm(v2))

print("Dot Product:", dot)
print("Cosine Similarity:", cos_theta)
```

```python
# Create grid
x = np.linspace(-3, 3, 10)
y = np.linspace(-3, 3, 10)
X, Y = np.meshgrid(x, y)

# Original grid
plt.figure()
plt.quiver(X, Y, X, Y, color='gray')
plt.title("Original Space")
plt.show()

# Transformation matrix
A = np.array([[2, 1],
              [1, 1]])

# Apply transformation
X_new = A[0,0]*X + A[0,1]*Y
Y_new = A[1,0]*X + A[1,1]*Y

plt.figure()
plt.quiver(X, Y, X_new, Y_new, color='blue')
plt.title("Transformed Space (Matrix Multiplication)")
plt.show()
```

```python
# Eigenvalues and Eigenvectors
eig_vals, eig_vecs = np.linalg.eig(A)

print("Eigenvalues:\n", eig_vals)
print("Eigenvectors:\n", eig_vecs)

# Plot eigenvectors
plt.figure()
plt.quiver(0, 0, eig_vecs[0,0], eig_vecs[1,0], color='red', scale=1)
plt.quiver(0, 0, eig_vecs[0,1], eig_vecs[1,1], color='green', scale=1)

plt.grid()
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.title("Eigenvectors of Matrix A")
plt.show()
```

```python
# Simple linear regression using normal equation
X = np.array([[1, 1], [1, 2], [1, 3]])
y = np.array([2, 4, 6])

# Normal equation
theta = np.linalg.inv(X.T @ X) @ X.T @ y

print("Model weights:", theta)
```
